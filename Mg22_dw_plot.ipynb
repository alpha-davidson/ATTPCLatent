```python
import matplotlib.pyplot as plt

# with pre-training
x1 = [175, 371, 521, 2171]
y1 = [0.85, 0.92, 0.93, 0.95]

# without pre-training
x2 = [175, 371, 521, 2171]
y2 = [0.45, 0.79, 0.90, 0.93]

# Create the plot
plt.plot(x1, y1, label='With Pre-Training', marker='o')  # Add markers to each point
plt.plot(x2, y2, label='Without Pre-Training', marker='s')  # Add markers to each point

# Add titles and labels
plt.title('Mg22 Track Counting Comparison')
plt.xlabel('Number of Events in Training Set')
plt.ylabel('Weighted Average F1-Score')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
