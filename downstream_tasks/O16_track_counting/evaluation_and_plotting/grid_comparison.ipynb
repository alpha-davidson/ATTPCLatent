```python
import os
from matplotlib import pyplot as plt
from PIL import Image

def create_grid(image_files, output_path, title):
    """
    Create a 3x3 grid of images and save it to a file.

    :param image_files: List of image file paths to include in the grid.
    :param output_path: Path to save the resulting grid.
    :param title: Title for the grid.
    """
    n_images = len(image_files)
    n_rows, n_cols = 3, 3  # Fixed 3x3 grid
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 10))

    for i in range(n_rows * n_cols):
        row, col = divmod(i, n_cols)
        if i < n_images:
            img = Image.open(image_files[i])
            axes[row, col].imshow(img)
            axes[row, col].axis('off')
            axes[row, col].set_title(os.path.basename(image_files[i]), fontsize=8)
        else:
            axes[row, col].axis('off')  # Turn off extra axes

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the title
    plt.savefig(output_path)
    plt.close(fig)

# Paths to the mislabeled and correct event plots
mislabeled_folder = '2024-12-10-00:12:13/mislabeled'
correct_folder = '2024-12-10-00:12:13/correct'

# Filter for the first 9 mislabeled files (event_0 to event_8)
mislabeled_files = sorted([f for f in os.listdir(mislabeled_folder) if f.startswith('mislabeled_event_') and 'true_3_pred_' in f])
mislabeled_files = [os.path.join(mislabeled_folder, f) for f in mislabeled_files if int(f.split('_')[2]) < 9]

# Filter for the first 9 correct files (event_0 to event_8)
correct_files = sorted([f for f in os.listdir(correct_folder) if f.startswith('correct_event_') and f.endswith('.png')])
correct_files = [os.path.join(correct_folder, f) for f in correct_files if int(f.split('_')[2]) < 9]

# Generate the 3x3 grid for mislabeled events
create_grid(
    mislabeled_files,
    '2024-12-10-00:12:13/mislabeled_grid.png',
    title='3x3 Grid of Mislabeled Events'
)

# Generate the 3x3 grid for correct events
create_grid(
    correct_files,
    '2024-12-10-00:12:13/correct_grid.png',
    title='3x3 Grid of Correctly Labeled Events'
)
```
