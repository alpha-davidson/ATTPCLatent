```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
true_labels_path = "../data_processing/data_splits/O16_size512_test_labels.npy"
y_true = np.load(f'{true_labels_path}')

predictions_path = "../training/benchmark_predictions/2024-12-14-03:13:09/O16_size512_predictions.npy"
y_predictions = np.load(f'{predictions_path}')
```

```python
y_prediction_labels = np.argmax(y_predictions, axis=1)

accuracy = accuracy_score(y_true, y_prediction_labels)
print(f"Accuracy: {accuracy}")
```

```python
cm = confusion_matrix(y_true, y_prediction_labels)
print(cm)
```

```python
report = classification_report(y_true, y_prediction_labels)
print(report)
```
