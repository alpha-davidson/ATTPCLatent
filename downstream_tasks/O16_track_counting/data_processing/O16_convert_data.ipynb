```python
file_name = ISOTOPE + '_w_key_index'
event_data = np.zeros((length, np.max(event_lens), 13), float)

for n in tqdm.tqdm(range(length)):
    name = keys[n]
    event = data[name]
    ev_len = len(event)
    for i,e in enumerate(event):
        instant = np.array(list(e))
        event_data[n][i][:12] = np.array(instant)
        event_data[n][i][-1] = float(n)
np.save('../voxel_data/' + file_name, event_data)
```
